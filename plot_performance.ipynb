```python
import csv
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_sgd = pd.read_csv('FC_SGD_stats.csv')
df_sgd
```

```python
df_sgd_mo = pd.read_csv('FC_SGD_Momentum_stats.csv')
df_sgd_mo
```

```python
df_sgd_bn = pd.read_csv('FC_BN_SGD_stats.csv')
df_sgd_bn
```

```python
df_adam = pd.read_csv('FC_Adam_stats.csv')
df_adam
```

## Training Time

```python
# Train time
tt_sgd = df_sgd['Training Time'][0]
tt_sgd_mo = df_sgd_mo['Training Time'][0]
tt_sgd_bn = df_sgd_bn['Training Time'][0]
tt_adam = df_adam['Training Time'][0]

tt_data = {
    'Algorithm': ['SGD', 'Momentum SGD', 'Momentum SGD with Batch Norm', 'Adam'],
    'Training Time': [tt_sgd, tt_sgd_mo, tt_sgd_bn, tt_adam]
}

df_train_time = pd.DataFrame(tt_data)

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='Algorithm', y='Training Time', data=df_train_time, palette='viridis')
plt.title('Comparison of Training Times by Optimization Algorithm')
plt.ylabel('Training Time (seconds)')
plt.xlabel('Algorithm')
plt.show()
```

## Training loss, validation loss, and final test loss 

```python
# Plotting
plt.figure(figsize=(8, 5)) 
sns.lineplot(data=df_sgd, x='Epoch', y='Training Loss', label='Training Loss')
sns.lineplot(data=df_sgd, x='Epoch', y='Validation Loss', label='Validation Loss')
sns.lineplot(data=df_sgd, x='Epoch', y='Test Loss', label='Test Loss')
plt.title('Loss Over Epochs: Neural Network Trained with SGD')
plt.xlabel('Epoch')
plt.ylabel('Loss')
```

```python
# Plotting
plt.figure(figsize=(8, 5)) 
sns.lineplot(data=df_sgd_mo, x='Epoch', y='Training Loss', label='Training Loss')
sns.lineplot(data=df_sgd_mo, x='Epoch', y='Validation Loss', label='Validation Loss')
sns.lineplot(data=df_sgd_mo, x='Epoch', y='Test Loss', label='Test Loss')
plt.title('Loss Over Epochs: Neural Network Trained with Momentum SGD')
plt.xlabel('Epoch')
plt.ylabel('Loss')
```

```python
# Plotting
plt.figure(figsize=(8, 5)) 
sns.lineplot(data=df_sgd_bn, x='Epoch', y='Training Loss', label='Training Loss')
sns.lineplot(data=df_sgd_bn, x='Epoch', y='Validation Loss', label='Validation Loss')
sns.lineplot(data=df_sgd_bn, x='Epoch', y='Test Loss', label='Test Loss')
plt.title('Loss Over Epochs: Neural Network Trained with Momentum SGD and Batch Normalization')
plt.xlabel('Epoch')
plt.ylabel('Loss')
```

```python
# Plotting
plt.figure(figsize=(8, 5)) 
sns.lineplot(data=df_adam, x='Epoch', y='Training Loss', label='Training Loss')
sns.lineplot(data=df_adam, x='Epoch', y='Validation Loss', label='Validation Loss')
sns.lineplot(data=df_adam, x='Epoch', y='Test Loss', label='Test Loss')
plt.title('Loss Over Epochs: Neural Network Trained with Adam')
plt.xlabel('Epoch')
plt.ylabel('Loss')
```
